# Libraries 


```python
# Import needed libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
```

# Loading The Dataset


```python
data= data = pd.read_csv('assets/DATA/SuperMarketSales.csv',  parse_dates=['Date'], dayfirst=True)
print(data.info())  
#Conclusion: 6435 entries and there isn't a null value to be imputed
```

```python
print(data.head())
```

```python
# some information about the values in  date
print(data['Date'].value_counts())
```

# Extract New Featues From Date Column 


```python
# Extract New Featues From Date Column (Year , Month , Day)
data['Date'] = pd.to_datetime(data['Date'])
data['Year'] = pd.DatetimeIndex(data['Date']).year
data['Month'] = pd.DatetimeIndex(data['Date']).month
data['Day'] = pd.DatetimeIndex(data['Date']).day

```

```python
# Delete The Date Column after we create the new feature.
data.drop(['Date'], axis=1, inplace=True)
```

# New DataSet

```python
print(data)
```

# dictionary To save Feature and its MSE

```python
# dictionary To save Feature and its MSE
my_dict = {}
```

# Store Feature

```python
# Store Feature
X_store = data["Store"]
y = data['Weekly_Sales']

X = np.expand_dims(X_store, axis=1)
Y = np.expand_dims(y, axis=1)
Simple_Linear_Regression = LinearRegression()
Simple_Linear_Regression.fit(X, Y)
Store_prediction = Simple_Linear_Regression.predict(X)
plt.scatter(X, Y)
plt.xlabel('Store', fontsize=20)
plt.ylabel('Weekly_Sales', fontsize=20)
plt.plot(X, Store_prediction, color='red', linewidth=3)
plt.show()
mse_Store = mean_squared_error(Y, Store_prediction)
my_dict["Store"] = mse_Store
print("Store", "Mean squared error:", mse_Store)
```
